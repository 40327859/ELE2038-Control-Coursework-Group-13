```python
import control as ctrl
import matplotlib.pyplot as plt
import numpy as np
from sympy import exp, Symbol
```

```python
#Impulse Plot no pid
s = ctrl.TransferFunction.s
G = 4562.9988239068/(s**3 + 15174.9010197085*s**2 + 243531.996259953*s - 3169857.09321053)

t_imp, x_imp = ctrl.impulse_response(G)
plt.grid()
plt.plot(t_imp, x_imp)
plt.title('Impulse Response Plot')
plt.xlabel("Time /s")
plt.ylabel("Distance /m")
plt.show()
```

```python
#Step Plot no pid
s = ctrl.TransferFunction.s
G = 4562.9988239068/(s**3 + 15174.9010197085*s**2 + 243531.996259953*s - 3169857.09321053)
t_imp, x_imp = ctrl.step_response(G)
plt.grid()
plt.plot(t_imp, x_imp)
plt.title('Step Response Plot')
plt.xlabel("Time /s")
plt.ylabel("Distance /m")
plt.show()
```

```python
#PID + Laser Both Impulse plot
G = 4562.9988239068/(s**3 + 15174.9010197085*s**2 + 243531.996259953*s - 3169857.09321053)
Kp = 1135
Kd = 22.5
Ki = 0

Glaser = 1 / (0.03 + 1)
controller = Kp + Ki/s + Kd*s
Gl = ctrl.feedback(G,controller)
Gfin = ctrl.feedback(Gl, Glaser)
t_imp, g_imp = ctrl.impulse_response(Gfin)

plt.plot(t_imp,g_imp*1000)
plt.grid()
plt.title('PID with Laser Impulse plot')
plt.xlabel("Time /s")
plt.xlabel("Time /s")
plt.ylabel("Distance /mm")
```

```python
#PID with Laser Step plot
G = 4562.9988239068/(s**3 + 15174.9010197085*s**2 + 243531.996259953*s - 3169857.09321053)
Kp = 1135
Kd = 22.5
Ki = 0

Glaser = 1 / (0.03 + 1)
controller = Kp + Ki/s + Kd*s
Gl = ctrl.feedback(G,controller)
Gfin = ctrl.feedback(Gl, Glaser)
t_imp1, g_imp1 = ctrl.step_response(Gfin)

plt.plot(t_imp1,g_imp1*1000)
plt.grid()
plt.title('PID with Laser Step plot')
plt.xlabel("Time /s")
plt.ylabel("Distance /mm")
plt.show()
```

```python
#PID + Laser Both Step and Impulse plot
G = 4562.9988239068/(s**3 + 15174.9010197085*s**2 + 243531.996259953*s - 3169857.09321053)
Kp = 1135
Kd = 22.5
Ki = 0

Glaser = 1 / (0.03 + 1)
controller = Kp + Ki/s + Kd*s
Gl = ctrl.feedback(G,controller)
Gfin = ctrl.feedback(Gl, Glaser)
t_imp, g_imp = ctrl.impulse_response(Gfin)
t_imp1, g_imp1 = ctrl.step_response(Gfin) 

plt.plot(t_imp,g_imp*1000)
plt.plot(t_imp1,g_imp1*1000)
plt.grid()
plt.title('PID with Laser sensor showing both step and impulse')
plt.legend(['Impulse Respnse', 'Step Response'])
plt.xlabel("Time /s")
plt.ylabel("Distance /mm")
plt.show()
```
